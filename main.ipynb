```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import sys
```

```python
#each line from each file reomve first line, if last character of line is ; remove it
#create directory if not exist with name new_ringoptimalcirculant1
if not os.path.exists("new_ringoptimalcirculant1"):
    os.makedirs("new_ringoptimalcirculant1")
for file in os.listdir("ringoptimalcirculant1"):
    if not file.endswith(".csv") or file[0] == "_":
        continue
    with open(os.path.join("ringoptimalcirculant1", file), "r") as f:
        new_file = open(os.path.join("new_ringoptimalcirculant1", file), "w")
        lines = f.readlines()
        f.seek(0)
        for line in lines[1:]:
            print(line)
            if line[-2] == ';':
                new_file.write(line[:-2] + "\n")
            else:
                new_file.write(line)

            new_file.flush()

print("os.listdir('ringoptimalcirculant1')[-3]", os.listdir("ringoptimalcirculant1")[-3])
```

```python
#Use normalized set of files
dfs = []
for file in os.listdir("new_ringoptimalcirculant1"):
    if not file.endswith(".csv") or file[0] == "_":
        continue
    dfs.append(pd.read_csv(os.path.join("new_ringoptimalcirculant1", file), delimiter=";"))
df = pd.concat(dfs, axis=0) # df columns
df = df.drop(columns=df.columns[[3]])
# Change columns names to Signature, diameter, average distance, number of connections
df.columns = ["Signature", "diameter", "average distance", "number_of_connections"]

df
```

```python
#save df to file
df.to_csv("new_ringoptimalcirculant1.csv", sep=";", index=False)
```

```python
#iterate over average distance column and replace "," with "." if "," in string
list = []
for i in df["average distance"]:
    if "," in str(i):
        r = float(str(i).replace(",", "."))
        list.append(r)
    else:
        r = float(i)
        list.append(r)
print(list)
df["average distance"] = list
```
